**1. Make a calculator using Python with addition, subtraction, multiplication, division and power**

```python
print("Calculator App")
print("---------------")

while True:
    print("Available Operators: (+, -, *, /, **) ")

    choice = input("Enter operator: ")

    n1 = float(input("Enter your fist number: "))
    n2 = float(input("Enter your second number: "))

    if choice == '+':
        result = n1 + n2
        print("Result: ", result)
    elif choice == '_':
        result = n1 - n2
        print("Result: ", result)
    elif choice == '*':
        result = n1 * n2
        print("Result: ", result)
    elif choice == '/':
        if n2 != 0:
            result= n1 / n2
            print("Result: ", result)
        else:
            print("Error! Division by zero is not allowed")
    elif choice == '**':
        result = n1**n2
        print("Result: ", result)
    else:
        print("Invalid operator")
        
    again = input("\nDo you want to calculate again? (yes/y)")
    again = again.strip().lower()

    if again not in ["yes", "y"]:
        print("Calculator closed!")
        break
    
```

**4. Write a Python program to sum all the numeric items in a dictionary**

```python
house = {
    "Rooms": 6,
    "Color" : "Brown and white",
    "kitchen" : 2,
    "Gallery" : 1,
}
total = 0
for key, value in house.items():
    if isinstance(value, (int, float)):
        total += value
print(total)
```

**5. Write a program to identify duplicate values from list.**

```python
my_list = [1, 5, 5, 7, 8 ,9, 9, 10, 15, 77]

seen = set()
duplicates = set()

for item in my_list:
    if item in seen:
        duplicates.add(item)
    else:
        seen.add(item)
print("Duplicate items are: ", duplicates)
    
```
