```python
print("Solution of Question no 7")
print("--------------------------")

Eng = int(input("Enter your obtained marks of English: "))
Urdu = int(input("Enter your obtained marks of Urdu: "))
Maths = int(input("Enter your obtained marks of Maths: "))
Chem = int(input("Enter your obtained marks of Chemistry: "))
Physics = int(input("Enter your obtained marks of Physics: "))

total = Eng + Urdu + Maths + Chem + Physics
perc = total / 500 * 100

print("Your percentage is: ", perc)

if perc>=90 and perc<=100:
    print("Grade A+")
elif perc>=80 and perc<=90:
    print("Grade A-1")
elif perc>=70 and perc<=80:
    print("Grade A")
elif perc>=60 and perc<=70:
    print("Grade B")
elif perc>=50 and perc<=60:
    print("Grade C")
elif perc>=40 and perc<=50:
    print("Grade D")
elif perc>=30 and perc<=40:
    print("Grade E")
else:
    print("You're failed!")
```
