```python
print("Calculator App")
print("---------------")
print("Select Operator")
print("1. Addition(+)")
print("2. Subtraction(-)")
print("3. Multiplication(*)")
print("4. Division(/)")

choice = input("Enter choice (1/2/3/4): ")

n1 = float(input("Enter your fist number: "))
n2 = float(input("Enter your second number: "))

if choice == '1':
    result = n1 + n2
    print("Result", result)
elif choice == '2':
    result = n1 - n2
    print("Result", result)
elif choice == '3':
    result = n1 * n2
    print("Result", result)
elif choice == '4':
    if n2 != 0:
        result= n1 / n2
        print("Result", result)
    else:
        print("Error! Division by zero is not allowed")
else:
    print("Invalid choice! PLease select 1,2,3 or 4")
    
    
    
         
```
